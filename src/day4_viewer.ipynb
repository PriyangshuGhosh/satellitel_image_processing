```python
import rasterio
import matplotlib.pyplot as plt
import numpy as np

with rasterio.open("Stacked_bands.tif") as src:
    r = src.read(1)
    g = src.read(3)
    b = src.read(4)
    nir = src.read(2)
rgb_composite = np.dstack((r, g, b))
rgb_composite = np.clip(rgb_composite, 0, 1)

plt.figure(figsize=(20, 10))
plt.imshow(rgb_composite)
plt.title("True Color Composite (RGB)")
plt.axis('off')
plt.show()
```
